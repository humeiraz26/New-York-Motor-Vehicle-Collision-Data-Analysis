```python
df = pd.read_csv("/content/drive/MyDrive/MVC.csv")

```

```python
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])
df['YEAR'], df['MONTH'] = df['CRASH_DATE'].dt.year, df['CRASH_DATE'].dt.month
```

```python
df.fillna(0)
```

```python
# To drop the na values for vechicle makes
df=df.dropna(subset=['VEHICLE_MAKE'])
```

```python
df
```

```python
import re
# Function to clean the names
def Clean_names(vehicle_name):
#Regex to search for hyphen(-) in the name followed by any #characters repeated any number of times
    if re.search('\-.*', vehicle_name):
        # Extract the position of beginning of pattern
        pos = re.search('\-.*', vehicle_name).start()
        print (vehicle_name[:pos])
        return vehicle_name[:pos].strip()
    else:
        # if clean up needed return the same name
        print (vehicle_name)
        return vehicle_name.strip()

# Update the vehicle_make columns
df['VEHICLE_MAKE'] = df['VEHICLE_MAKE'].apply(Clean_names)
```

```python
df['VEHICLE_MAKE'] = df['VEHICLE_MAKE'].astype('str')

```

```python
df['YEAR'] = df['YEAR'].astype(str)
```

```python
df_2018=df[df['YEAR'].str.contains('2018')]
df_2019=df[df['YEAR'].str.contains('2019')]
df_2020=df[df['YEAR'].str.contains('2020')]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


carbrand = ['HOND','MERZ','ISU','SUBA']




```

```python

ac_2018 = {}
ac_2019 = {}
ac_2020 = {}
```

```python
for i in carbrand:
    ac_2018[i] = ac_2018.get(i,0)+df_2018['VEHICLE_MAKE'].value_counts()[i]
    ac_2019[i] = ac_2019.get(i,0)+df_2019['VEHICLE_MAKE'].value_counts()[i]
    ac_2020[i] = ac_2020.get(i,0)+df_2020['VEHICLE_MAKE'].value_counts()[i]
```

```python
barWidth = 0.30
br1 = np.arange(len(carbrand))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

#br1 is the initial point to plot the data, where np.arrange divides the length.
plt.bar(br1, ac_2018.values(), color ='r', width = barWidth,
        edgecolor ='grey', label ='2018')
plt.bar(br2, ac_2019.values(), color ='g', width = barWidth,
        edgecolor ='grey', label ='2019')
plt.bar(br3, ac_2020.values(), color ='b', width = barWidth,
        edgecolor ='grey', label ='2020')

plt.xticks(br1, carbrand)
plt.xlabel("Brands")
plt.ylabel("Number of accidents" )
plt.title("Accidents occured" )
plt.legend()
plt.show()
```
